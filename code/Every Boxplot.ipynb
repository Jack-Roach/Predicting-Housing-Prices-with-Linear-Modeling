### Imports

```python
#General Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import sys
if not sys.warnoptions:
    import warnings
    warnings.simplefilter("ignore")
```

### Reading in Data

```python
train = pd.read_csv("../datasets/train_cleaned.csv")
test = pd.read_csv("../datasets/test_cleaned.csv")
```

### Creating a list of categorical features from our dataset
#### This will be needed to quickly plot each feature

```python
num_features = [n for n in train._get_numeric_data().columns if n != 'SalePrice']
cat_features = [n for n in train.columns if n not in num_features and n != 'SalePrice']
print('Number of categorical features:',len(cat_features))
```

```python
cat_features
```

### Box Plots of Every Categorical Variable
#### We will choose ones with obvious category discrepencies to analyze further in the Visualization file.

```python
plt.figure(figsize=(10,15))
for n in cat_features:
    plt.figure(figsize=(10,15))
    sns.boxplot(y=train[n],x=train['SalePrice'])
    plt.title(n);
```
